```python
from flask import Flask, render_template_string
import requests

app = Flask(__name__)

# âœ… Your News API key
API_KEY = "7e63a1fa97274bfeb50f8c57b1bd2f66"

# HTML template written as a string (no separate file needed)
HTML_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>ðŸ“° Latest News</title>
    <style>
        body { font-family: Arial, sans-serif; background-color: #fafafa; padding: 20px; }
        .news-container { display: flex; flex-wrap: wrap; gap: 20px; }
        .news-card {
            background-color: white; padding: 15px; border-radius: 10px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1); width: 300px;
        }
        img { width: 100%; height: 180px; object-fit: cover; border-radius: 8px; }
        h3 { font-size: 18px; margin-bottom: 10px; }
        a { color: #007bff; text-decoration: none; }
    </style>
</head>
<body>
    <h1>ðŸ“° Latest News</h1>
    <div class="news-container">
        {% for article in articles %}
        <div class="news-card">
            <h3>{{ article['title'] }}</h3>
            <p>{{ article['description'] }}</p>
            {% if article['urlToImage'] %}
            <img src="{{ article['urlToImage'] }}" alt="News Image">
            {% endif %}
            <p><a href="{{ article['url'] }}" target="_blank">Read more</a></p>
        </div>
        {% endfor %}
    </div>
</body>
</html>
"""

@app.route('/',methods=['GET'])
def home():
    url = f"https://newsapi.org/v2/top-headlines?country=us&apiKey={API_KEY}"
    response = requests.get(url)
    data = response.json()

    if data.get('status') == 'ok':
        articles = data.get('articles', [])
    else:
        articles = []

    return render_template_string(HTML_TEMPLATE, articles=articles)
```

```python
from werkzeug.serving import run_simple

# Start Flask server (use Ctrl+C to stop it)
run_simple('localhost', 5000, app, use_reloader=False)
```
