```python
import cv2
from matplotlib import pyplot as plt

image_path = 'pic.jpg'
image = cv2.imread(image_path)
resized_image = cv2.resize(image, (1900, 800))
resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
plt.imshow(resized_image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
Gaussian = cv2.GaussianBlur(resized_image, (15, 15), 0)  
Gaussian_rgb = cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB)  
plt.imshow(Gaussian_rgb)
plt.title('Gaussian Blurred Image')
plt.axis('off')
plt.show()
```

```python
median = cv2.medianBlur(resized_image, 11)  
median_rgb = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)  

plt.imshow(median_rgb)
plt.title('Median Blurred Image')
plt.axis('off')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('pic2.jpg')


downsampled_image = cv2.pyrDown(image)
# Convert BGR → RGB before showing
downsampled_rgb = cv2.cvtColor(downsampled_image, cv2.COLOR_BGR2RGB)

plt.imshow(downsampled_rgb)
plt.title('Downsampled Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image_path = 'pic2.jpg'  

image = cv2.imread(image_path)

if image is None:
    print("Image not found! Check the path or file name.")
else:
    # Upsample the image
    upsampled_image = cv2.pyrUp(image)

    # Convert BGR to RGB for displaying correctly with matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    upsampled_rgb = cv2.cvtColor(upsampled_image, cv2.COLOR_BGR2RGB)

    # Show both images side by side
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(upsampled_rgb)
    plt.title("Upsampled Image (pyrUp)")
    plt.axis("off")

    plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load your image
image_path = 'pic2.jpg'  # Make sure the image is in the same folder as your notebook
image = cv2.imread(image_path)

if image is None:
    print("Image not found! Check the file name or path.")
else:
    # Get height and width
    height, width = image.shape[:2]
    quarter_height, quarter_width = height / 4, width / 4

    # Define the translation matrix
    T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

    # Apply translation
    img_translation = cv2.warpAffine(image, T, (width, height))

    # Convert BGR to RGB for correct display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    translated_rgb = cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB)

    # Display both images side by side
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(translated_rgb)
    plt.title("Translated Image")
    plt.axis("off")

    plt.show()
```
