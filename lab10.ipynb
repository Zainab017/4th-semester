```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.chat.util import Chat, reflections
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('vader_lexicon')
```

```python
# [q1][ans1]
# [q1,q2][ans1]
# [q1,q2][ans1,ans2]
pairs=[
    ["hello",   ["hi","heyy","hellooo"]],
    ["hi",      ["hi, how can i help u ? "]],
    ["who r u", ["i m just a chatbot, how can i help you?"]],
    [r"(?i).*degrees.*offer.*|tell me about the programs offered in superior university.*",#r(?i) u can add imp words
    ["BSAI,BSDS,BSCS,BSIT,BSSE"]]
]

```

```python
sia = SentimentIntensityAnalyzer
chatbot=Chat(pairs, reflections)
```

```python
def analyze_sentiments(text):
    score=sia.polarity_scores(text)
    # print(score)
    if score['compound']>=0.05:
      print("positive sentence")
    if score['compound']<=0.05:
      print("negative sentence")
    else:
      print("neutral")
```

```python
def chatbot_app():
    print("chatbot:hello i m your chatbot. ")
    while True:
        user_input=input("user: ")
        if user_input.lower()=="bye":
            print("goodbye, have a good day")
            break
        elif "sentiment" in user_input.lower():
            sentiment_input=input("user:")
            print("chatbot: sentiment analysis result",analyze_sentiments(sentiment_input))
            pass
        else:
            response=chatbot.respond(user_input)
            if response:
              print("chatbot:",response)
            else:
              print("chatbot: i m sorry ! i dont know")
              break
chatbot_app()
```
